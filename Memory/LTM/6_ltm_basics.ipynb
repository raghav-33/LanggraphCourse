# Memory Store

Now we will learn about Memory Store, how it is created , search 

After that we will learn how to Integrate Will LLM

```python
from langgraph.store.memory import InMemoryStore
# InMemory Store things in Ram , Not persistance But fast
```

```python
# creating a store  
store = InMemoryStore()
```

```python
# creating a namespace 

# namespace is like Folder in your google Drive in which we organize data
# lly memory store kai andar memory ko organize krne kai lai namespace use krte hai.
namespace = ("user", "u1")   # User is parent folder | inside it u1 folder
```

### Creating Memories

put() Method is Used

```python
# Adding memories
# put method  : put method in particular namespace create new memory 
# Basically used for creating new memorie
# Syntax (namespcae , key , value)
store.put(namespace, "1", {"data": "User likes pizza"})
store.put(namespace, "2", {"data": "User prefers dark mode"})
```

```python
# another namespace
namespace2 = ("user", "u2")  # another  child folder u2 inside root folder user

# Add memories
store.put(namespace2, "1", {"data": "User likes pasta"})
store.put(namespace2, "2", {"data": "User prefers grid style navigation"})
```

### Retrieving Memories

get() method is Used

```python
#  Syntax : store.get(namespace, key)
store.get(namespace2, "2")
```

### Retrieving all Memories

search() method is used 

```python
items = store.search(namespace2)

for item in items:
    print(item.value)
```

### Semantic Search

get(): Retrive One Memory only


search() : Retrive all Memory 

this method fail in certain situation
like we want to get only that memory in which user is talking about "Mumbai"

NOTE: Now we are searching Based on "Meaning"

```python
# Embedding Model 
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding_model = OpenAIEmbeddings(model='text-embedding-3-small')
```

```python
# Passing Embedding Model to store For semantic search
store = InMemoryStore(index={'embed': embedding_model, 'dims':1536})
```

```python
# Namespace Created (ROOT folder)
namespace = ('users', 'u1')
```

```python
# Creating Memories (Child Folder)
store.put(namespace, "1", {"data": "User prefers concise answers over long explanations"})
store.put(namespace, "2", {"data": "User likes examples in Python"})
store.put(namespace, "3", {"data": "User usually works late at night"})
store.put(namespace, "4", {"data": "User prefers dark mode in applications"})
store.put(namespace, "5", {"data": "User is learning machine learning"})
store.put(namespace, "6", {"data": "User dislikes overly theoretical explanations"})
store.put(namespace, "7", {"data": "User prefers step-by-step reasoning"})
store.put(namespace, "8", {"data": "User is based in India"})
store.put(namespace, "9", {"data": "User likes real-world analogies"})
store.put(namespace, "10", {"data": "User prefers bullet points over paragraphs"})
```

```python
# Semantic search performing
# passing : "query" For which related to retrive
# limit: maximum no of memory fetch . like kwargs in langchain 
items = store.search(namespace, query="what is the user currently learning", limit=1)

for item in items:
    print(item.value)
```

```python
items = store.search(namespace, query="what are user's preferences", limit=3)

for item in items:
    print(item.value)
```
