# Customer Review Workflow

Customer will give a review about product . we have to find sentiment of review whether it's positive or negative.

Then We run diagonsis if person is facing problem then due to which issue like bad UI , hang problem

if review is positive then run positive response operation and lly for negative response.

```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict, Literal
from dotenv import load_dotenv
from pydantic import BaseModel, Field

load_dotenv()
```

```python
model = ChatOpenAI(model='gpt-4o-mini')
```

```python
# Structured Output 1
class SentimentSchema(BaseModel):
    sentiment: Literal['positive' , 'Negative' , 'Neutral'] = Field(description = "sentiment of the Review")
```

```python
# Structured Output 2
class DiagnosisSchema(BaseModel):
    issue_type: Literal['UX' ,'Performance' , 'Bug' , 'Support' , 'other'] = Field(description='The category of issue mentioned in the review')
    tone: Literal["angry", "frustrated", "disappointed", "calm"] = Field(description='The emotional tone expressed by the user')
    urgency: Literal["low", "medium", "high"] = Field(description='How urgent or critical the issue appears to be')
```

```python
structured_model = model.with_structured_output(SentimentSchema)
structured_model2 = model.with_structured_output(DiagnosisSchema)
```

```python
# State Define
class ReviewState(TypedDict):
    review:str
    sentiment: Literal['positive' , 'negative' , 'netural']
    diagnosis: str
    response: str
```

```python
def find_sentiment(state: ReviewState):
    prompt = f'For the following review find out the sentiment \n {state["review"]}'
    sentiment = structured_model.invoke(prompt).sentiment
    return {'sentiment': sentiment}

def check_sentiment(state: ReviewState) -> Literal["positive_response", "run_diagnosis"]:

    if state['sentiment'] == 'positive':
        return 'positive_response'
    else:
        return 'run_diagnosis'


def positive_response(state: ReviewState):
    prompt = f"""Write a warm thank-you message in response to this review:
    \n\n\"{state['review']}\"\n
Also, kindly ask the user to leave feedback on our website."""
    response = model.invoke(prompt).content
    return {'response': response}

def run_diagnosis(state: ReviewState):
    prompt = f"""Diagnose this negative review:\n\n{state['review']}\n"
    "Return issue_type, tone, and urgency.
"""
    response = structured_model2.invoke(prompt)
    return {'diagnosis': response.model_dump()}


    def negative_response(state: ReviewState):
        diagnosis = state['diagnosis']
        prompt = f"""You are a support assistant.
The user had a '{diagnosis['issue_type']}' issue, sounded '{diagnosis['tone']}', and marked urgency as '{diagnosis['urgency']}'.
Write an empathetic, helpful resolution message.
"""
        response = model.invoke(prompt).content
        return {'response': response}
```

```python
# Graph Define
graph = StateGraph(ReviewState)

# Nodes
graph.add_node('find_sentiment', find_sentiment)
graph.add_node('positive_response', positive_response)
graph.add_node('run_diagnosis', run_diagnosis)
graph.add_node('negative_response', negative_response)

# Edges
graph.add_edge(START, 'find_sentiment')

graph.add_conditional_edges('find_sentiment', check_sentiment)

graph.add_edge('positive_response', END)

graph.add_edge('run_diagnosis', 'negative_response')
graph.add_edge('negative_response', END)

# Graph compile
workflow = graph.compile()

```
